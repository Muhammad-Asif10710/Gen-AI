```python
from transformers import AutoTokenizer
model="bert-base-uncased"
tokenizer=AutoTokenizer.from_pretrained(model)
text="i love asif, tokenize"
tokenized_text=tokenizer(text)
tokenized_text
```

Token type ids define as 0 Sentence A: The first sentence (or passage) in the input. as 1 Sentence B: The second sentence (or related text), which can either be a continuation or a separate context
